```python
import pandas as pd
import numpy as np
import seaborn as sns
#visualisation
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_theme(color_codes=True)
```

```python
df = pd.read_csv("cars_data.csv") # Loading the dataset
```

```python
#displaying first five rows
df.head()
```

```python
#displaying last five rows
df.tail()
```

```python
df.dtypes #checking data types of each column
```

```python
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1) #dropping unnecessary columns
df.head(6) #displaying first 6 rows
```

```python
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Total Cylinders", "Transmission Type": "Mechanism", "Driven_Wheels": "Drive Mode", "highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price"}) #renaming columns
df.head(7)
```

```python
df.shape #checking number of rows and columns (rows, columns)
```

```python
df.count() # Used to count the number of rows
```

```python
duplicate_rows_df = df[df.duplicated()] #checking duplicate rows
print("number of duplicate rows:", duplicate_rows_df.shape) 
```

```python
df = df.drop_duplicates() #dropping duplicate rows
df.head(5) 
```

```python
print(df.isnull().sum()) #checking for null values in each column
```

```python
df = df.dropna() #dropping null values  
df.count() 
```

```python
# Visualizing the distribution of Price column using boxplot
sns.boxplot(x=df['Price'])
plt.show()
```

```python
# Visualizing the distribution of Total Cylinders column using boxplot
sns.boxplot(x=df['HP'])
plt.show()
```

```python
# Visualizing the distribution of Total Cylinders column using boxplot
sns.boxplot(x=df['Total Cylinders'])
plt.show()
```

```python
#Identify numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns
#Compute Q1, Q3, and IQR for only numeric columns
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1
# Filter rows that are NOT outliers in any numeric column
# Keep only rows where all numeric values are within the IQR bounds
condition = ~((df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)
# Apply this condition to the full Dataframe (all columns)
df_cleaned = df[condition]
# Check the new shape
print(df_cleaned.shape)
```

```python
# Scatter plot between HP and Price
fig, ax = plt.subplots(figsize=(10,6)) # setting the figure size
ax.scatter(df['HP'], df['Price'])  
ax.set_xlabel('HP')
ax.set_ylabel('Price')
plt.show()
```

```python
# Bar plot showing number of cars for each Make
df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars Vs Make")
plt.ylabel('Number of cars')
plt.xlabel('Make')
plt.show()
```

```python
# Correlation heatmap
#First, filter only numeric columns for correlation
numeric_df = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(10,5))
c = numeric_df.corr() # Calculate correlation only on numeric columns
sns.heatmap(c, cmap="BrBG", annot=True)
# Uncomment to display the plot
plt.show()
```
